Создать объект `Series` с индексами `a, b, c` и значениями `1, 2, 3` соответственно:
- из списка;
- из массива `NumPy`;
- из словаря.

```python
import pandas as pd
import numpy as np

s1 =  pd.Series([1,2,3], ['a', 'b', 'c'])
print(s1)
s2 = pd.Series(np.array([1,2,3]), np.array(['a', 'b', 'c']))
print(s2)
s3 = pd.Series({'a': 1, 'b' : 2, 'c' : 3})
print(s3)
```

```python
df[df['age'] == 31][['region', 'sum']]
print(df.loc[(df['age'] == 31), ['region', 'sum']])
```

Выбрать людей в возрасте от 30 до 40 лет.

```python
print(df[df['age'].between(30, 40)])
print(df[ (df['age'] >= 30) & (df['age'] < 40) ])
print(df.loc[((df['age'] >= 30) & (df['age'] < 40))])
```

Выбрать людей в возрасте меньше 30 лет, проживающих в Курске или Орле.

```python
print(df[ (df['age'] < 30) & ( (df['region'] == 'Орёл') | (df['region'] == 'Курск') )])
print(df.loc[(df['age'] < 30) & (df['region'].isin(['Орёл', 'Курск']))])
```

Выбрать людей, которые не проживают в Москве.

# **4. Операции со строками и столбцами**

В таблицу `df` добавить новый столбец `'payment'` (способ оплаты), принимающий следующие значения Sber Pay, cash, QIWI Кошелёк, bank card, QR-код, cash, Sber Pay, bank card, bank card, Sber Pay.  

```python
df =df.assign(payment=['Sber Pay', 'cash', 'QIWI Кошелёк', 'bank card', 'QR-код', 'cash', 'Sber Pay', 'bank card', 'bank card', 'Sber Pay'])
```

Переименовать столбец `'num'` в `'purchase'`.

```python
df = df.rename(columns={'num': 'purchase'})
df
```

Создать и добавить в таблицу `df` новый стобец `'average'`, содержащий среднюю стоимость покупок.

1. Решить задачу с использованием метода `apply()` и lambda-функции.

```python
df['average'] = df.apply(lambda row: row['sum'] / row['purchase'], axis =1 )
df
```

Удалить столбец `'average'`.

```python
df = df.drop('average', axis=1)
df
```

Решить задачу с использованием функции `.map()`.

Создать и добавить в таблицу `df` новый стобец `'rating'`, содержащий рейтинг покупателей в зависимости от средней стоимости покупок:
- 1 - средняя стоимость покупок меньше < 500;
- 2 - средняя стоимость покупок >= 500 и < 1000;
- 3 - средняя стоимость покупок >= 1000.


```python
df['average_cost'] = df['sum'] / df['purchase']
def set_rating(average_cost):
  if average_cost < 500:
    return "1"
  elif average_cost < 1000:
    return "2"
  else:
    return "3"

df['rating'] = df['average_cost'].map(set_rating)
print(df)
```
